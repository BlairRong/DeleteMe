Task 2: Add extraction functions



```python
# Task 2 – Extract specific information (numbers, total, date)

import re

def extract_numbers(text):
    """Return all numbers (integers and decimals) found in the text."""
    return re.findall(r'\d+\.?\d*', text)

def find_total(lines):
    """Look for a line containing 'TOTAL' and extract the number after it."""
    for line in lines:
        if 'TOTAL' in line.upper():
            # Try to find the first number in that line
            numbers = extract_numbers(line)
            if numbers:
                return numbers[0]  # return the first number (assumed total)
    return None

def find_date(lines):
    """Look for a line containing a date pattern (MM/DD or MM/DD/YY)."""
    date_pattern = r'\b\d{1,2}/\d{1,2}(?:/\d{2,4})?\b'
    for line in lines:
        matches = re.findall(date_pattern, line)
        if matches:
            return matches[0]
    return None
```

Task 4 – Improve with Prompting (Text Cleaning & Readability)



Step 1 – Add a cleaning function


The function will:

Remove excessive newlines.

Strip leading/trailing spaces from each line.

Filter out lines that consist only of symbols (e.g., ><, ---).

Replace multiple spaces with a single space.

Optionally remove stray characters that are not letters, numbers, or common punctuation.

```python
import re

def clean_text(text):
    """Clean OCR output: normalize spaces, remove noise, and filter empty lines."""
    # Replace multiple newlines with a single newline
    text = re.sub(r'\n\s*\n', '\n', text)
    # Remove lines that are mostly non‑alphanumeric (e.g., decoration lines)
    lines = []
    for line in text.splitlines():
        # Remove leading/trailing spaces
        line = line.strip()
        # Keep only lines with at least one alphanumeric character
        if re.search(r'[a-zA-Z0-9]', line):
            # Replace multiple spaces inside the line
            line = re.sub(r' +', ' ', line)
            lines.append(line)
    return '\n'.join(lines)
```

Task 5 – Add Simple Logic (Expense Categorization)

I extend the OCR application to automatically detect the total amount from a receipt and classify it as High expense (> 500) or Normal expense (≤ 500).

```python
Step 1 – Create a function to extract the total amount

I look for a line containing the word TOTAL (or TOT, Total) and then extract the first number in that line.
```

```python
def extract_total(text):
    """
    Extract the total amount from OCR text.
    Looks for lines containing 'TOTAL', then extracts the first number.
    Returns the total as a float, or None if not found.
    """
    for line in text.splitlines():
        if 'TOTAL' in line.upper():
            numbers = re.findall(r'\d+\.?\d*', line)
            if numbers:
                return float(numbers[0])
    return None
```

Step 2 – Add the logic to the main program

After cleaning the text and extracting numbers, I call extract_total() and print the category.

```python
# Task 5 – OCR with expense categorization

!apt-get install -y tesseract-ocr
!pip install pytesseract pillow

import pytesseract
import re
from PIL import Image
from google.colab import files

def clean_text(text):
    """Clean OCR output: normalize spaces, remove noise, filter empty lines."""
    text = re.sub(r'\n\s*\n', '\n', text)
    lines = []
    for line in text.splitlines():
        line = line.strip()
        if re.search(r'[a-zA-Z0-9]', line):
            line = re.sub(r' +', ' ', line)
            lines.append(line)
    return '\n'.join(lines)

def extract_total(text):
    """Extract the total amount (first number in a line containing 'TOTAL')."""
    for line in text.splitlines():
        if 'TOTAL' in line.upper():
            numbers = re.findall(r'\d+\.?\d*', line)
            if numbers:
                return float(numbers[0])
    return None

# Upload image
print("Please upload an image of a receipt:")
uploaded = files.upload()
filename = next(iter(uploaded))
print(f"\nFile uploaded: {filename}")

# OCR
img = Image.open(filename)
raw_text = pytesseract.image_to_string(img)

# Clean text
cleaned_text = clean_text(raw_text)

# Extract total
total = extract_total(cleaned_text)

# Output
print("\n" + "="*50)
print("CLEANED EXTRACTED TEXT")
print("="*50)
print(cleaned_text)

print("\n" + "="*50)
print("EXPENSE CATEGORY")
print("="*50)
if total is None:
    print("Could not find total amount.")
else:
    print(f"Total: {total:.2f}")
    if total > 500:
        print("High expense")
    else:
        print("Normal expense")
```

**Where does OCR stop and “understanding” begin?**  
OCR ends when it converts visual text into machine‑readable strings. Understanding begins when we interpret that text – e.g., identifying which number is the total, what the date means, or that “TOTAL” signals the final amount. The separation is between **extraction** (OCR) and **interpretation** (parsing, context awareness).

**Did my system actually understand the receipt?**  
No, it did not understand. It simply applied rules: it looked for a line containing the word “TOTAL” and took the first number. It has no concept of what a receipt is, what subtotal vs. total means, or that “TOTAL” might be misread. It performed pattern matching, not understanding.

**What would be needed to make it smarter?**  
- **Contextual knowledge**: Know that receipts often have “SUBTOTAL” followed by “TOTAL”, so it should look for the second occurrence.  
- **Better parsing**: Use more robust regex or a small ML model to identify key fields (total, date, tax) based on layout and proximity.  
- **Fallback mechanisms**: Handle OCR errors (e.g., “T0TAL” or “TOT AL”) using fuzzy matching.  
- **Document structure awareness**: Distinguish between line items, headers, and summary sections.  
- **Domain adaptation**: Train a small model on receipt layouts, or use a pre‑built service like Azure Document Intelligence that understands document structure.

Task 7 — Advanced: Multi‑Image Receipt Analyzer with Field Extraction


In this task I extend the OCR system to:

Detect if an image contains a receipt (simple keyword‑based detection).

Extract specific fields – total amount and date.

Process multiple images in one run, showing results for each.

All code runs in Google Colab.

```python
# Task 7 – Advanced Multi‑Image Receipt Analyzer

# Install dependencies
!apt-get install -y tesseract-ocr
!pip install pytesseract pillow

import pytesseract
import re
from PIL import Image
from google.colab import files

def clean_text(text):
    """Remove extra whitespace and keep only lines with alphanumeric chars."""
    text = re.sub(r'\n\s*\n', '\n', text)
    lines = []
    for line in text.splitlines():
        line = line.strip()
        if re.search(r'[a-zA-Z0-9]', line):
            line = re.sub(r' +', ' ', line)
            lines.append(line)
    return '\n'.join(lines)

def is_receipt(text):
    """Detect if the text likely comes from a receipt."""
    keywords = ['TOTAL', 'SUBTOTAL', 'RECEIPT', 'TAX', 'CASH', 'CHANGE', 'ITEM']
    upper_text = text.upper()
    matches = sum(1 for kw in keywords if kw in upper_text)
    return matches >= 2

def extract_total(text):
    """Find total amount (looks for 'TOTAL' line and extracts first number)."""
    for line in text.splitlines():
        if 'TOTAL' in line.upper():
            numbers = re.findall(r'\d+\.?\d*', line)
            if numbers:
                return float(numbers[0])
    return None

def extract_date(text):
    """Find a date in common formats (MM/DD or MM/DD/YY)."""
    patterns = [r'\b\d{1,2}/\d{1,2}/\d{2,4}\b', r'\b\d{1,2}/\d{1,2}\b']
    for line in text.splitlines():
        for pat in patterns:
            match = re.search(pat, line)
            if match:
                return match.group()
    return None

def process_image(filename):
    img = Image.open(filename)
    raw = pytesseract.image_to_string(img)
    cleaned = clean_text(raw)
    return {
        'filename': filename,
        'receipt': is_receipt(cleaned),
        'total': extract_total(cleaned),
        'date': extract_date(cleaned),
        'preview': cleaned[:150].replace('\n', ' ')
    }

# Upload multiple images
print("Upload one or more images (hold Ctrl/Cmd to select multiple):")
uploaded = files.upload()
filenames = list(uploaded.keys())
print(f"\nUploaded {len(filenames)} file(s): {filenames}")

# Process each
results = [process_image(f) for f in filenames]

# Print summary
print("\n" + "="*60)
print("RECEIPT ANALYSIS SUMMARY")
print("="*60)
for r in results:
    print(f"\n📄 {r['filename']}")
    if r['receipt']:
        print("   ✅ Likely a receipt")
        print(f"   💰 Total: {r['total'] if r['total'] else 'not found'}")
        print(f"   📅 Date:  {r['date'] if r['date'] else 'not found'}")
    else:
        print("   ❌ Not a receipt (or keywords missing)")
    print(f"   👀 Preview: {r['preview']}")
```
